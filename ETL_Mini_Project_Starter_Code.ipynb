### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
columns = crowdfunding_info_df.columns

columns
```

```python
# Split 'category & sub-category' column into 'category' and 'subcategory' columns using '/'
split_categories = crowdfunding_info_df['category & sub-category'].str.split('/', n=1, expand=True)

# Assign the split columns to the DataFrame
crowdfunding_info_df['category'] = split_categories[0]
crowdfunding_info_df['subcategory'] = split_categories[1]

# Handle cases where no delimiter is found
no_delimiter_mask = split_categories[1].isna()
crowdfunding_info_df.loc[no_delimiter_mask, 'subcategory'] = None

# Drop the original 'category & sub-category' column
crowdfunding_info_df.drop(columns=['category & sub-category'], inplace=True)

# Print the DataFrame to verify changes
print(crowdfunding_info_df)
```

```python
# Get unique categories and subcategories
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()

# Print the unique categories and subcategories
print("Unique Categories:")
print(categories)

print("\nUnique Subcategories:")
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# List comprehension to add "cat" to each category ID
cat_ids = [f"cat{id}" for id in categories]

# List comprehension to add "subcat" to each subcategory ID
scat_ids = [f"subcat{id}" for id in subcategories]

# Print the resulting lists
print(cat_ids)
print(scat_ids)
```

```python
import pandas as pd

# Create a DataFrame for categories
category_df = pd.DataFrame({
    'category': cat_ids,
    'category_name': categories
})

# Create a DataFrame for subcategories
subcategory_df = pd.DataFrame({
    'subcategory': scat_ids,
    'subcategory_name': subcategories
})

```

### Create the Contacts DataFrame 
---
**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

Then export the DataFrame as a `contacts.csv` CSV file.

### Option 1: Use Pandas to create the contacts DataFrame.

### Option 2: Use regex to create the contacts DataFrame.

```python
contact_info_df_copy = contact_info_df.copy()
contact_info_df_copy.head()
```

```python
# Assuming the column name is currently spaces or empty strings
contact_info_df_copy = contact_info_df_copy.rename(columns={'   ': 'contact_info'})

# Now you can proceed with your operations on the 'contact_info' column

# Extract the four-digit contact ID number.
import re

# Assuming the contact information is stored in a column named 'contact_info' in the DataFrame
contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_info'].str.extract(r'"contact_id": (\d{4})')

# Remove the 'four_digit_contact_id' column
contact_info_df_copy.drop(columns=['four_digit_contact_id'], inplace=True)


contact_info_df_copy.head()
```

```python
# Check the datatypes.
contact_info_df_copy.dtypes
```

```python
#Drop NaaN values
contact_info_df_copy = contact_info_df_copy.dropna(subset=['contact_id'])
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_info'].str.extract(r'"contact_id":\s*(\d+)').astype('int64')
```

```python
# Extract the name of the contact and add it to a new column.
contact_info_df_copy['name'] = contact_info_df_copy['contact_info'].str.extract(r'"name": "([^"]*)"')

# Display the DataFrame with the new 'name' column
contact_info_df_copy.head()
```

```python
# Extract the email from the contacts and add the values to a new column.
contact_info_df_copy['email'] = contact_info_df_copy['contact_info'].str.extract(r'"email": "([^"]*)"')

# Display the DataFrame with the new 'email' column
contact_info_df_copy.head()
```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contact_info_subset_df = contact_info_df_copy[['contact_id', 'name', 'email']].copy()

# Display the new DataFrame with selected columns
contact_info_subset_df.head()
```

```python
# Extracting first and last names using regex
contact_info_df_copy[['first_name', 'last_name']] = contact_info_df_copy['name'].str.extract(r'^(\w+)(?:\s+(\w+))?$')

# Dropping the original "name" column
contact_info_df_copy.drop(columns=['name'], inplace=True)

# Display the updated DataFrame
contact_info_df_copy.head()
```

```python
# Define the regex pattern
pattern = r'(first_name|last_name|email|contact_id)'

# Filter columns based on the regex pattern
columns_order = [col for col in contact_info_df_copy.columns if re.search(pattern, col)]

# Reorder the DataFrame columns
contact_info_df_copy = contact_info_df_copy[columns_order]

# Display the updated DataFrame
contact_info_df_copy.head()

```
